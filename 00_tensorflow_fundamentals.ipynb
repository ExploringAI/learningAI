Introduction to Tensors
If you've ever used NumPy, tensors are kind of like NumPy arrays (we'll see more on this later).

For the sake of this notebook and going forward, you can think of a tensor as a multi-dimensional numerical representation (also referred to as n-dimensional, where n can be any number) of something. Where something can be almost anything you can imagine:

It could be numbers themselves (using tensors to represent the price of houses).
It could be an image (using tensors to represent the pixels of an image).
It could be text (using tensors to represent words).
Or it could be some other form of information (or data) you want to represent with numbers.
The main difference between tensors and NumPy arrays (also an n-dimensional array of numbers) is that tensors can be used on GPUs (graphical processing units) and TPUs (tensor processing units).

The benefit of being able to run on GPUs and TPUs is faster computation, this means, if we wanted to find patterns in the numerical representations of our data, we can generally find them faster using GPUs and TPUs.

**Matrix multiplication**

In machinge learning, matrix multiplication is one of the most common tensor operations.

There are two rules needed to matrix multiply:
* the dimensions must match
* the resulting matrix has the shape of the outer dimensions


```python
# Matrix mult in TensorFlow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
# Matrix mult with Python operation "@"
tensor @ tensor
```

```python
# Create a tensor of (3,2)
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
# Create another (3,2) tensor
Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X,Y
```

```python
# Try to matrix pultiply tensors of same shape
tf.matmul(X,Y)
```

```python
# Change the shape of Y:
tf.reshape(Y, shape=(2,3))
```

```python
# Create a tensor of (3,2)
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
# Create different (2,3) tensor
Y = tf.constant([[7,8,9],
                 [10,11,12]])
X,Y
```

```python
# Try to matrix pultiply tensors of compatable shape
tf.matmul(X,Y)
```

```python
# Can do the same with transpose (But it flips the axies)
tf.transpose(X), tf.reshape(X,shape=(2,3))
```

```python
# Try matrix mult with transpose
tf.matmul(tf.transpose(X), tf.transpose(Y))
```

**The dot product**
Matrix mult is also referred as the dot product.

You can preform matriz mult using:
* `tf.matmul()`
* `tf.tensordot`

```python
Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X, Y
```

```python
# Preform dot product on X & Y (requires X or Y to be transposed)
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Preform matrix mult between X & Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Preform matrix mult between X & Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
# check the vals of Y, reshape Y and transposed Y
print("Normal Y:")
print(Y, "\n") # \n is newline

print("Y reshaped to (2,3):")
print(tf.reshape(Y, (2,3)))

print(" \nY transposed:")
print(tf.transpose(Y))
```

```python
tf.matmul(X, tf.transpose(Y))
```

Usually, when doing mult on 2 tensors and 1 of the axes doesn't line up, you will tranpose (rather than reshape) one of the tensors to satisfy matrix mult rules.

### Aggregating tensors

Aggregating tensors = condensing them down from many values to fewer values

```python
# Get the absolute values
M = tf.constant([-7,-10])
M
```

```python
tf.abs(M)
```

Lets go through thr following forms of aggregation:
* Get min
* Get max
* Get mean
* Get sum

```python
# Create a random tensor with values between 0 and 100 
E = tf.constant(np.random.randint(0,100,size=50))
E
```

```python
# Find the min
tf.reduce_min(E)
```

```python
# Find the max
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the variance of our tensor, we need tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# or .math
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

```python
# Find the standard deviation with tfp
tfp.stats.stddev(tf.cast(E, dtype=tf.float32))
```

```python
# Find the standard deviation with tf
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```
